Visualization Graphs

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Year",data=kollywood)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="Year",data=kollywood)
```

```python
sns.barplot(x="movie_rating",y="Hero_Rating",data=kollywood)
plt.title("kollywood movie")
plt.xlabel("movie_rating")
plt.ylab("Hero_Rating")
plt.show()
```

```python
sns.lineplot(x="PeopleVote",y="movie_rating",data=kollywood,marker='o',label='Movie_Rating',color='blue')
```

```python
sns.scatterplot(x='Year',y="movie_rating",data=kollywood ,s=100,color='red')
plt.xlabel("Year")
plt.ylabel("movie_rating")
plt.title("Scatter Plot")
     
```
